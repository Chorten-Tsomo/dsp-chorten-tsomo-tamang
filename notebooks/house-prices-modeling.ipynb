```python
import pandas as pd
import numpy as np
```

```python
#Load the data
train_data = pd.read_csv("C:\\Users\\tsomo\\dsp-chorten-tsomo-tamang\\data\\house-prices-advanced-regression-techniques\\train.csv")
test_data = pd.read_csv("C:\\Users\\tsomo\\dsp-chorten-tsomo-tamang\\data\\house-prices-advanced-regression-techniques\\test.csv")
```

```python
train_data.head(20)
```

```python
#Seperate data into features (X) and target (y)
X = train_data.drop(columns=['SalePrice'])
y = train_data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#Handling missing data
print(train_data.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer
#Impute missing categorical value with the mode
imputer = SimpleImputer(strategy='most_frequent')
X_train_imputed = imputer.fit_transform(X_train)
```

```python
#Feature Selection
continuous_features = ['LotArea', 'YearBuilt']
categorical_features = ['Neighborhood', 'ExterQual']
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Initialize encoders and scalers
scaler = StandardScaler()
encoder = OneHotEncoder(handle_unknown='ignore')

# Fit encoders and scalers to training data
scaler.fit(X_train[continuous_features])
encoder.fit(X_train[categorical_features])

# Transform continuous features
X_train_scaled = scaler.transform(X_train[continuous_features].copy())
X_test_scaled = scaler.transform(X_test[continuous_features].copy())

# One-hot encode categorical features
X_train_encoded = encoder.transform(X_train[categorical_features].copy()).toarray()
X_test_encoded = encoder.transform(X_test[categorical_features].copy()).toarray()

# Merge processed features
X_train_processed = np.concatenate((X_train_scaled, X_train_encoded), axis=1)
X_test_processed = np.concatenate((X_test_scaled, X_test_encoded), axis=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_processed, y_train)
```

```python
#Predict on the test set
y_pred = model.predict(X_test_processed)
```

```python

from sklearn.metrics import mean_squared_log_error



def compute_rmsle(y_test: np.ndarray, y_pred: np.ndarray, precision: int = 2) -> float:
    rmsle = np.sqrt(mean_squared_log_error(y_test, y_pred))
    return round(rmsle, precision)

#Compute RMSE
rmsle = compute_rmsle(y_test, y_pred)
print("Root Mean Squared Log Error(RMSE):", rmsle)
```

```python
# Convert processed features arrays into dataframes
X_train_processed_data = pd.DataFrame(X_train_processed, columns=continuous_features + list(encoder.get_feature_names_out(categorical_features)))
X_test_processed_data = pd.DataFrame(X_test_processed, columns=continuous_features + list(encoder.get_feature_names_out(categorical_features)))
```

```python
# Save the processed dataframes to Parquet files
X_train_processed_data.to_parquet('/my/filepath/X_train_processed.parquet', index=False)
X_test_processed_data.to_parquet('/my/filepath/X_test_processed.parquet', index=False)
```
