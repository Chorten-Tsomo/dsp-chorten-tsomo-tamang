```python
import pandas as pd
import numpy as np
```

```python
#Load the data
train_data = pd.read_csv("C:\\Users\\tsomo\\dsp-chorten-tsomo-tamang\\data\\house-prices-advanced-regression-techniques\\train.csv")
test_data = pd.read_csv("C:\\Users\\tsomo\\dsp-chorten-tsomo-tamang\\data\\house-prices-advanced-regression-techniques\\test.csv")
```

```python
#Seperate data into features (X) and target (y)
X = train_data.drop(columns=['SalePrice'])
y = train_data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#Handling missing data
print(train_data.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer
#Impute missing categorical value with the mode
imputer = SimpleImputer(strategy='most_frequent')
X_train_imputed = imputer.fit_transform(X_train)
```

```python
#Feature Selection
continuous_features = ['LotArea', 'YearBuilt']
categorical_features = ['Neighborhood', 'ExterQual']
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler

#Processing steps for continuous features 
scaler = StandardScaler()
X_train_scaled = X_train[continuous_features].copy()
X_test_scaled = X_test[continuous_features].copy()
X_train_scaled = scaler.fit_transform(X_train_scaled)
X_test_scaled = scaler.transform(X_test_scaled)


#One hot encode categorical features
encoder = OneHotEncoder(handle_unknown='ignore')
X_train_encoded = X_train[categorical_features].copy()
X_test_encoded = X_test[categorical_features].copy()
X_train_encoded = pd.get_dummies(X_train_encoded, columns=categorical_features, drop_first=True)
X_test_encoded = pd.get_dummies(X_test_encoded, columns=categorical_features, drop_first=True)

# Merge processed features
X_train_processed = np.concatenate((X_train_scaled, X_train_encoded.values), axis=1)
X_test_processed = np.concatenate((X_test_scaled, X_test_encoded.values), axis=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_processed, y_train)
```

```python
#Predict on the test set
y_pred = model.predict(X_test_processed)
```

```python

from sklearn.metrics import mean_squared_log_error



def compute_rmsle(y_test: np.ndarray, y_pred: np.ndarray, precision: int = 2) -> float:
    rmsle = np.sqrt(mean_squared_log_error(y_test, y_pred))
    return round(rmsle, precision)

#Compute RMSE
rmsle = compute_rmsle(y_test, y_pred)
print("Root Mean Squared Log Error(RMSE):", rmsle)
```
